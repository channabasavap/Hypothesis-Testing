Null_Hypothesis=There is no significant difference between cutlet samples from two outlet: Mu1=Mu2

Alternate_Hypothesis=There is significant difference between cutlet samples from two outlet: Mu1!=Mu2

Alpha1=5%=0.05

```python
import pandas as pd
import scipy.stats as stats
```

```python
cutlets=pd.read_csv("C:\\Users\\chann\\Downloads\\Cutlets.csv")
```

```python
cutlets.head(10)
```

```python
Alpha1=0.05
T_test_Value=stats.ttest_ind(a=cutlets['Unit A'],b=cutlets['Unit B'])
```

```python
P1 = T_test_Value[1]
print("P1:", P1)

if P1 <= 0.05:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

Here we are taking 4 samples and comparing so use ANOVA TEST FOR THIS.

Null_Hypothesis=All samples are related(MEAN OF ALL SAMPLES ARE EQUAL)

Alternate_Hypothesis=All samples are not related(MEAN OF ALL SAMPLES ARE not EQUAL)

Alpha=0.05

```python
LabTAT=pd.read_csv("C:\\Users\\chann\\Downloads\\LabTAT.csv")
```

```python
LabTAT
```

```python
LabTAT.info()
```

```python
#Prforming ANOVA TESTING
Alpha=0.05
Anova=stats.f_oneway(LabTAT.iloc[:,0],LabTAT.iloc[:,1],LabTAT.iloc[:,2],LabTAT.iloc[:,3])
Anova
```

```python
if Anova[1]<=Alpha:
  print("Accept Alternate Hypothesis,So that All the datas are not related each other")
else:
  print("Failed to Reject Null Hypothesis,So that All the datas are related each other")
```
